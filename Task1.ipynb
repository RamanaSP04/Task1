List operation-ADDITION

```python
my_list = [1, 2, 3]
my_list.append(4)
print(my_list)
```

List operation-DELETION

```python
my_list.remove(2)
print(my_list)
```

List operation-MODIFICATION

```python
my_list[1] = 5
print(my_list)
```

Dictionary operation-ADDITION

```python
my_dict = {'a': 1, 'b': 2}
my_dict['c'] = 3
print(my_dict)
```

Dictionary operation-DELETION

```python
del my_dict['b']
print(my_dict)
```

Dictionary operation-MODIFICATION

```python
my_dict['a'] = 10
print(my_dict)
```

Set operation-ADDITION

```python
my_set = {1, 2, 3}
my_set.add(4)
print(my_set)
```

Set operation-DELETION

```python
my_set.remove(2)
print(my_set)
```

Set operation-Modification

```python
my_set.remove(3)
my_set.add(5)
print(my_set)
```
